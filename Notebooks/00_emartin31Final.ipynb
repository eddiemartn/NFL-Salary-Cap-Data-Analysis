```python
## Websites Used https://www.nfl.com/standings/league/2023/REG and https://www.spotrac.com/nfl/cap/_/year/2023/sort/cap_maximum_space
### The NFL.com website tracks the each teams record as well as some statistics such as points scored and allowed.
### The spotrac.com website tracks the financials for each team such as payroll.
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
# This block of code scrapes the data from Spotrac.com and adds it to a data frame
driver = webdriver.Chrome()

all_teams = []
all_cap_space = []
all_years = []
def spotrac_data(year):
    url = f"https://www.spotrac.com/nfl/cap/_/year/{year}/sort/cap_maximum_space"
    driver.get(url)

    team = []
    cap_space = []

    for i in range(1,33):
        team_name_xpath = f'//*[@id="table-wrapper"]/div[2]/div[2]/table/tbody/tr[{i}]/td[2]'
        cap_amount_xpath = f'//*[@id="table-wrapper"]/div[2]/div[2]/table/tbody/tr[{i}]/td[7]'

        team_name = driver.find_element(By.XPATH, team_name_xpath).text
        cap_amount = driver.find_element(By.XPATH, cap_amount_xpath).text

        team.append(team_name)
        cap_space.append(cap_amount)

    return team, cap_space

for year in range(2023, 2010, -1):
    print(f"Scraping data for year {year}...")
    teams, cap_space = spotrac_data(year)
    all_teams.extend(teams)
    all_cap_space.extend(cap_space)
    all_years.extend([year] * len(teams))

data = {
    'Year': all_years,
    'Team Abbr': all_teams,
    'Cap Space': all_cap_space
}

cap_df = pd.DataFrame(data)

print(cap_df)
cap_df.to_csv('Spotrac_Raw_Data.csv', index = False)
driver.quit()
```

```python
#This block of code scrapes data from the NFL and puts it into a data frame
driver = webdriver.Chrome()

all_teams = []
all_wins = []
all_losses = []
all_ties = []
all_win_pcts = []
all_years = []

def nfl_data(year):
    url = f"https://www.nfl.com/standings/league/{year}/REG"
    driver.get(url)

    teams = []
    wins = []
    losses = []
    ties = []
    win_pct = []

    for i in range(1,33):
        team_name_xpath = f'//*[@id="main-content"]/section[3]/div/div/div/div/div[2]/table/tbody/tr[{i}]/td[1]'
        team_wins_xpath = f'//*[@id="main-content"]/section[3]/div/div/div/div/div[2]/table/tbody/tr[{i}]/td[2]'
        team_losses_xpath = f'//*[@id="main-content"]/section[3]/div/div/div/div/div[2]/table/tbody/tr[{i}]/td[3]'
        team_ties_xpath = f'//*[@id="main-content"]/section[3]/div/div/div/div/div[2]/table/tbody/tr[{i}]/td[4]'
        team_pct_xpath = f'//*[@id="main-content"]/section[3]/div/div/div/div/div[2]/table/tbody/tr[{i}]/td[5]'

        team_name = driver.find_element(By.XPATH, team_name_xpath).text
        team_wins = driver.find_element(By.XPATH, team_wins_xpath).text
        team_losses = driver.find_element(By.XPATH, team_losses_xpath).text
        team_ties = driver.find_element(By.XPATH, team_ties_xpath).text
        team_win_pct = driver.find_element(By.XPATH, team_pct_xpath).text

        team_name = team_name.replace("\nx\nz", "").replace("\nx", "").replace("\n*", "").replace("\ny","")
        
        teams.append(team_name)
        wins.append(team_wins)
        losses.append(team_losses)
        ties.append(team_ties)
        win_pct.append(team_win_pct)

    return teams, wins, losses, ties, win_pct

for year in range(2023,2010,-1):
    print(f"Scraping data for year {year}...")
    teams, wins, losses, ties, win_pct = nfl_data(year)
    all_teams.extend(teams)
    all_wins.extend(wins)
    all_losses.extend(losses)
    all_ties.extend(ties)
    all_win_pcts.extend(win_pct)
    all_years.extend([year] * len(teams))

data = {
    'Year': all_years,
    'Team': all_teams,
    'Wins': all_wins,
    'Losses': all_losses,
    'Ties': all_ties,
    'PCT': all_win_pcts
}

nfl_df = pd.DataFrame(data)

print(nfl_df)
nfl_df.to_csv('NFL_Raw_Data.csv', index = False)
driver.quit
```

```python
# This gives the team names that have changed over this time period a common name
nfl_df['Team'] = nfl_df['Team'].replace({
    'Redskins': 'Commanders',
    'Football Team': 'Commanders',
    'Niners': '49ers'
})

print(nfl_df)
```

```python
# This combines both data frames into one
merged_df = pd.merge(nfl_df, cap_df, on = ['Team' , 'Year'], how = 'left')
merged_df = merged_df.drop(columns = ['Team Abbr'])
merged_df
merged_df.to_csv('NFL_Cap_Data_Raw.csv', index = False)
```
